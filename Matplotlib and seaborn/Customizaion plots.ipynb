# ***Customizing Plots in Matplotlib and Seaborn***


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\PC\OneDrive\Desktop\Data\diamonds.csv')
df
```

```python
print(f'number of rows: {df.shape[0]}')
print(f'number of columns: {df.shape[1]}')
print(f'{50*'-'}')
df=df.sample(5000)
print(f'number of rows after sampling : {df.shape[0]}')
print(f'number of columns after sampling : {df.shape[1]}')

df.head()
      
```

```python
sns.scatterplot(x='carat',y='price',data=df,hue='color')
plt.legend(title='color', loc='upper right')
plt.title('Scatter plot of carat vs price')
plt.xticks(rotation=45, ha='right',fontsize= 10)
plt.xlabel('carat')
plt.ylabel('price', labelpad=10)
plt.show()
```

```python
plt.scatter(x=df['carat'], y=df['price'])
plt.title('Scatter plot of carat vs price')
plt.xlabel('carat')
plt.ylabel('price')

plt.show()
```
